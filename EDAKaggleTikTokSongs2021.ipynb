Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Read file

```python
df = pd.read_csv("https://raw.githubusercontent.com/vinaykumara09/Kaggle/main/TikTok_songs_2021x.csv")
df
```

Sort by artist's popularity

```python
sorted_by_popularity = df.sort_values(by=['artist_pop'], ascending = False)
sorted_by_popularity
```

Plot by artist's popularity

```python
plt.subplots(figsize = (30, 10))
plt.xticks(rotation = 90)
sns.barplot(x="artist_name", y="artist_pop", data = sorted_by_popularity, ci = "sd")
plt.show()
```

```python
sorted_by_danceability = df.sort_values(by=['danceability'], ascending = False)
sorted_by_danceability
```
